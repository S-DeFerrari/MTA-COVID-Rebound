```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns


df = pd.read_csv('2019 Sales Feb - July - Sheet1.csv')
```

```python
df
```

```python
df.plot(x='day_number')
```

```python
for index, row in df.iterrows():
    print(index, row['amount_collected'])
```

```python
df.describe()
```

```python
low_ball = df.loc[(df['amount_collected'] < 1657.545000)]

low_ball.groupby(['month']).count().sort_values('day_number')
```

```python
day_of = df.groupby(['month','day_of_Week']).amount_collected.mean().reset_index()

day_of.sort_values(['month','amount_collected'],ascending=False).reset_index(drop=True)
```

```python
ftcombo = pd.read_csv('bedford 2020-19 nums.csv')
```

```python
ftcombo
```

```python
yy = ftcombo.iloc[:,3]
xx = ftcombo.iloc[:,2]

yy2 = ftcombo.iloc[:,4]
```
